```python

from datetime import date, time  

def insert_test_slots():
    conn = db.get_connection()
    cursor = conn.cursor()

    # Example test slots (change provider_id, type_id as per your DB)
    test_slots = [
        (1, date(2025, 9, 25), "09:00", "10:00", "available", 1),
        (1, date(2025, 9, 25), "10:00", "11:00", "available", 1),
        (2, date(2025, 9, 25), "11:00", "12:00", "booked", 1),
        (2, date(2025, 9, 25), "14:00", "15:00", "available", 1),
        (1, date(2025, 9, 25), "15:00", "16:00", "available", 1),
    ]

    insert_query = """
        INSERT INTO availability_slots (provider_id, date, start_time, end_time, status, type_id)
        VALUES (%s, %s, %s, %s, %s, %s)
    """

    try:
        cursor.executemany(insert_query, test_slots)
        conn.commit()
        print(f"Inserted {cursor.rowcount} test slots successfully.")
    except Exception as e:
        conn.rollback()
        print("Error inserting slots:", e)
    finally:
        conn.close()
```

```python
insert_test_slots()
```
